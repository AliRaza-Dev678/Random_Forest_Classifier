## Project: Random Forest Tree Classifier(Iris Dataset)

```python
# Importing/Loading Dataset
import pandas as pd 
df = pd.read_csv('Iris.csv')
df.head()
```

```python
# Splitting Dataset into Input and Output variables
x = df[['Id', 'SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
y = df['Species']
```

```python
x.head()
```

```python
y.head()
```

```python
# Splitting Dataset into Training and Testing Sets (80/20)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
# Creating and Fitting model 
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100).fit(x, y)
model
```

```python
# Checking Predicted values on x_test
predicted_values = model.predict(x_test)
predicted_values
```

```python
# Checking Accuracy Score
from sklearn.metrics import accuracy_score
score = accuracy_score(predicted_values, y_test)
print(score*100,'%')
```

```python
# Checking Accuracy by Confusion Matrix
from sklearn import metrics
cm = metrics.confusion_matrix(predicted_values, y_test)
cm
```

```python
# Making Graph of Accuracy of model 
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 6))
sns.heatmap(cm, annot=True, fmt='.3f', linewidths=.5, square=True, cmap='Spectral')
plt.ylabel('Actual Output')
plt.xlabel('Predicted Output')
plt.title('Accuracy Score: 100.0%', size = 15)
plt.show()
```
